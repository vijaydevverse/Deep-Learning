 **Deep Learning (DL)**
 ===


**Deep Learning** is a method that allows computers to learn from large amounts of data and make predictions. It's called **deep** learning because it uses a structure of layers (like layers of a cake) to process information. The idea is that by stacking many layers, a deep learning model can understand very complex patterns, like recognizing objects in a photo, understanding speech, or translating languages.

 **Artificial Neural Networks (ANN)**
 ===

At the heart of deep learning is something called **Artificial Neural Networks (ANNs)**, which are inspired by how the human brain works. The brain has billions of neurons, and ANNs are a simplified version of this system.

1. **Neurons and Layers**
   - **Neurons** are the basic units of an ANN. Each neuron takes an input, processes it, and gives an output.
   - **Layers** in an ANN are groups of neurons that work together. You can think of it like a team where each member does a small task, and together they solve a bigger problem.
     - **Input Layer**: This is the first layer that receives the data (like an image or a number).
     - **Hidden Layers**: These layers perform calculations to process the data and extract important features (like edges in an image).
     - **Output Layer**: This layer gives the final result or prediction (for example, it might predict if an image is a cat or a dog).

   **Example**:  
   - If the input is an image of a cat, the network will first look at the raw pixel data, then extract features like shapes and colors in the hidden layers, and finally in the output layer, it will decide that the image is a "cat."

2. **Activation Functions**
   - **Activation Functions** decide if a neuron should "fire" or activate. They determine how much influence a neuronвЂ™s result should have on the next layer. Some common ones are:
     - **ReLU (Rectified Linear Unit)**: It passes positive values as they are and makes negative values zero. This helps the network learn faster and better.
     - **Sigmoid**: It turns values into a range between 0 and 1, which is often used when making predictions about probabilities (like yes/no questions).
     - **Tanh**: This one squashes the values between -1 and 1 and is sometimes used to make the network learn more efficiently.

   **Example**:  
   - Imagine a neuron calculates a value of 5. With **ReLU**, it would pass the value 5 on to the next layer, but if the value was -3, it would become 0 and not pass it on.

3. **Backpropagation**
   - After the network makes a prediction, the prediction is compared to the actual value (like comparing the predicted cat image to the real label "cat").
   - **Backpropagation** is a method that helps the network learn from its mistakes. It adjusts the neurons to reduce the error. The error is "sent backward" through the network, and the weights (the strength of the connections between neurons) are adjusted.

   **Example**:  
   - If the network incorrectly classifies a cat as a dog, backpropagation helps it learn from the mistake and adjust its internal settings for future predictions.

4. **Gradient Descent**
   - **Gradient Descent** is a process that helps the network find the best settings (or weights) for its neurons. The network starts with random weights and keeps adjusting them based on how well itвЂ™s performing. It tries to minimize the error by making small changes to the weights.
   - ItвЂ™s like a hiker trying to find the lowest point in a valley (the "best" solution). The hiker takes small steps downhill, adjusting direction as they go, until they reach the lowest point.

   **Example**:  
   - In training, the network keeps making predictions, calculates the error, and then uses gradient descent to adjust the weights to reduce the error step by step.

5. **Loss Function**
   - A **Loss Function** measures how far off the network's prediction is from the actual answer. The goal is to minimize this loss (error) as much as possible.
     - For example:
       - **Mean Squared Error (MSE)** is often used for predicting continuous values (like predicting house prices).
       - **Cross-Entropy Loss** is used for classification tasks (like predicting if an image is a cat or a dog).

   **Example**:  
   - If the network predicts a house price as $300,000, but the actual price is $350,000, the loss function would calculate the difference, and the network would learn to adjust to minimize that difference in future predictions.

---

 **Training Process**
 ===

Training a deep learning model involves repeatedly showing it data (like pictures or numbers) and adjusting it to improve its predictions.

- **Step 1**: Data is fed into the network (input layer).
- **Step 2**: The network processes the data through the layers and makes a prediction (output layer).
- **Step 3**: The prediction is compared to the actual result (e.g., real labels of cat/dog images).
- **Step 4**: The error (difference) between the predicted and actual result is calculated using the loss function.
- **Step 5**: The error is sent backward through the network (backpropagation) to adjust the weights.
- **Step 6**: **Gradient Descent** is used to minimize the error by adjusting the weights in small steps.
- **Step 7**: Repeat the process (this is done many times, called "epochs"), and the network keeps improving its predictions.

---

 **In Summary (Simplified)**
 ===

Deep Learning uses networks of neurons (like our brain) to recognize patterns in data, such as images or text. It involves:
1. **Feeding data** to the network (e.g., pictures of cats).
2. **Processing data** in layers to extract important features.
3. **Making predictions** and comparing them to the actual results.
4. **Adjusting the network** through backpropagation and gradient descent to reduce errors.
5. **Improving over time** by repeating the process with many examples (this is called "training").

Deep Learning helps computers get better and better at tasks like recognizing faces, understanding speech, and translating languages by learning from vast amounts of data.

#classification

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Generating synthetic data
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Preprocessing: Scaling the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Defining the ANN model
model = Sequential()
model.add(Dense(64, input_dim=20, activation='relu'))  # Hidden Layer 1
model.add(Dense(32, activation='relu'))               # Hidden Layer 2
model.add(Dense(1, activation='sigmoid'))             # Output Layer

# Compiling the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Training the model
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
# Test prediction
# Need to add a batch dimension to the single sample X_test[0]
probs = model.predict(np.expand_dims(X_test[0], axis=0))

pred_classes = np.round(probs).astype(int)

print("Probabilities:\n", probs.flatten())
print()
print("Predicted classes (by rounding):", pred_classes.flatten())
```

```python
# Evaluating the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
def remove_outliers_iqr(df,exclude_column):
  for col in df.columns:
    if col==exclude_column:
      continue
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR

    df=df[(df[col]>=lower_bound) & (df[col]<=upper_bound)]
  return df

df_cleaned=remove_outliers_iqr(df,exclude_column='price')
df=df_cleaned
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from itertools import product
```

```python
#set the param
plt.rc('figure', autolayout=True)
plt.rc('image', cmap='magma')
```

```python
# define the kernel
kernel = tf.constant([[-1, -1, -1],
                    [-1,  8, -1],
                    [-1, -1, -1],
                   ])
```

```python
#load the image
image=tf.io.read_file('/content/Ganesh.jpg')
image=tf.io.decode_jpeg(image,channels=1)
image=tf.image.resize(image,size=[300,300])
```

```python
#plot the image
img=tf.squeeze(image).numpy()
plt.figure(figsize=(5,5))
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.title('Original Gray Scale image')
plt.show();
```

```python
#Reformat
image=tf.image.convert_image_dtype(image,dtype=tf.float32)
image=tf.expand_dims(image,axis=0)
kernel=tf.reshape(kernel,[*kernel.shape,1,1])
kernel=tf.cast(kernel,dtype=tf.float32)
```

```python
#Convolution layer
conv_fn=tf.nn.conv2d

image_filter=conv_fn(
    input=image,
    filters=kernel,
    strides=1, # or (1,1)
    padding='SAME',
)

plt.figure(figsize=(15,5))
```

```python
#Plot the convolved image
plt.subplot(1,3,1)

plt.imshow(
    tf.squeeze(image_filter)
)
plt.axis('off')
plt.title('Convolution')
```

```python
#Activation Layer
relu_fn=tf.nn.relu
```

```python
#Image Detection
image_detect=relu_fn(image_filter)
```

```python
plt.subplot(1,3,2)
plt.imshow(
    #Reformat for plotting
    tf.squeeze(image_detect)
)
plt.axis('off')
plt.title('Activation')
```

```python
#Pooling Layer
pool=tf.nn.pool
image_condense=pool(input=image_detect,
                    window_shape=(2,2),
                    pooling_type='MAX',
                    strides=(2,2),
                    padding='SAME',
                    )
plt.subplot(1,3,3)
plt.imshow(tf.squeeze(image_condense))
plt.axis('off')
plt.title('Pooling')
plt.show()
```

###With Plotting

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np

# 1. Load and Preprocess the CIFAR-10 dataset
(x_train,y_train),(x_test,y_test)=datasets.cifar10.load_data()
x_train,x_test=x_train/255.0,x_test/255.0

# Class names for CIFAR-10
class_names=['Airplane','Automobile','Bird','Cat','Deer',
               'Dog','Frog','Horse','Ship','Truck']

# 2. Build the CNN model
model=models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128,(3,3),activation='relu'),

    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(10,activation='softmax')
])

# 3. Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 4. Train the model
history=model.fit(x_train,y_train,epochs=40,
                    validation_data=(x_test,y_test))

# 5. Evaluate the model
test_loss,test_acc=model.evaluate(x_test,y_test)
print(f"\nTest accuracy: {test_acc:.4f}")

# 6. Predict on test images
predictions=model.predict(x_test)

# 7. Display predictions for 5 random test images
num_images=5
plt.figure(figsize=(10,5))

for i in range(num_images):
    index=np.random.randint(0,len(x_test))
    plt.subplot(1,num_images,i + 1)
    plt.xticks([]); plt.yticks([]); plt.grid(False)
    plt.imshow(x_test[index])

    predicted_label=class_names[np.argmax(predictions[index])]
    true_label=class_names[int(y_test[index])]

    color='green' if predicted_label==true_label else 'red'
    plt.xlabel(f"P: {predicted_label}\nT: {true_label}",color=color)

plt.tight_layout()
plt.show()
```

###Recurrent Neural Networks

<b>Step 1: Import Necessary Libraries</b> <br>

We start by importing essential libraries for data handling and building the neural network.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

<b>Step 2: Define the Input Text and Prepare Character Set </b> <br>
We define the input text and identify unique characters in the text which weвЂ™ll encode for our model.

```python
text="This is GreekforGeeks a software training institute"
chars=sorted(list(set(text)))
char_to_index={char: i for i, char in enumerate(chars)}
index_to_char={i: char for i, char in enumerate(chars)}
```

```python
chars
```

```python
char_to_index
```

```python
index_to_char
```

**Step 3: Create Sequences and Labels** <br>
To train the RNN, we need sequences of fixed length (seq_length) and the character following each sequence as the label.

```python
seq_length=3
sequences=[]
labels=[]

for i in range(len(text) - seq_length):
  seq=text[i:i + seq_length]
  label=text[i + seq_length]
  sequences.append([char_to_index[char] for char in seq])
  labels.append(char_to_index[label])

X=np.array(sequences)
y=np.array(labels)
```

```python
sequences
```

```python
labels
```

**Step 4: Convert Sequences and Labels to One-Hot Encoding** <br>
For training, we convert X and y into one-hot encoded tensors.

```python
X_one_hot=tf.one_hot(X, len(chars))
y_one_hot=tf.one_hot(y, len(chars))
```

**Step 5: Build the RNN Model**

We create a simple RNN model with a hidden layer of 50 units and a Dense output layer with softmax activation.

```python
model=Sequential()
model.add(SimpleRNN(50, input_shape=(seq_length, len(chars)), activation='relu'))
model.add(Dense(len(chars), activation='softmax'))
```

**Step 6: Compile and Train the Model**

We compile the model using the categorical_crossentropy loss and train it for 100 epochs.

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(X_one_hot,y_one_hot,epochs=100)
```

**Step 7: Generate New Text Using the Trained Model**

After training, we use a starting sequence to generate new text character-by-character.

```python
start_seq="this is G"
generated_text=start_seq

for i in range(50):
  x=np.array([[char_to_index[char] for char in generated_text[-seq_length:]]])
  x_one_hot=tf.one_hot(x, len(chars))
  prediction=model.predict(x_one_hot)
  next_index=np.argmax(prediction)
  next_char=index_to_char[next_index]
  generated_text += next_char

print("Generated Text:")
print(generated_text)
```
